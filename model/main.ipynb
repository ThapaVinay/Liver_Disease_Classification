```python
import pandas as pd

df= pd.read_csv("data.csv")

df.head()
```

```python
df.shape
```

```python
df.Liver_Disease.value_counts()
```

```python
df['Alcohol_Consumption'] = df['Alcohol_Consumption'].map({
    'No' : 0,
    'Frequent' : 1, 
    'Sometimes' : 2
})
```

```python
df['Smoking_Status'] = df['Smoking_Status'].map({
    'No' : 0,
    'Frequent' : 1, 
    'Sometimes' : 2
})
```

```python
df.corr()
```

```python
import seaborn as sns

sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
X = df.drop("Liver_Disease", axis=1)
y = df['Liver_Disease']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=2024)
```

```python
X_train.shape
```

```python
X_test.shape
```

#### Creating model

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = knn.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
import numpy as np

def predict_disease(jaundice, fatigue, discomfort, appetite, urine, itchy_skin, bruising, nausea_vomiting, smoking_status, alcohol_consumption):
    
    x = np.zeros(len(X_train.columns))

    x[0] = jaundice
    x[1] = fatigue
    x[2] = discomfort
    x[3] = appetite
    x[4] = urine
    x[5] = itchy_skin
    x[6] = bruising
    x[7] = nausea_vomiting
    x[8] = smoking_status
    x[9] = alcohol_consumption

    return knn.predict([x])[0]
```

```python
predict_disease(1, 1, 1, 1, 1, 1, 1, 1, 0, 2)
```

#### Exporting the model

```python
import pickle
with open("liver_disease_prediction.pickle", "wb") as f:
    pickle.dump(knn, f)
```
